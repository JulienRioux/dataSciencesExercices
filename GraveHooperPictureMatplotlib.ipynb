```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
fig = plt.figure(figsize=(25, 10))

# read the image file
imgArr = plt.imread("Grace_Hopper.png")


# Create the subplot 
# red
ax1 = plt.subplot2grid((2,3), (0,0))
# green
ax2 = plt.subplot2grid((2,3), (0,1))
# blue
ax3 = plt.subplot2grid((2,3), (1,0))
# light
ax4 = plt.subplot2grid((2,3), (1,1))
# photo
ax5 = plt.subplot2grid((2,3), (0,2), rowspan=2)
ax5.axis("off")
# Show the image
ax5.imshow(imgArr)

# Set title for each subplot
ax1.set_title("Red", fontdict={"fontsize":24})
ax2.set_title("Green", fontdict={"fontsize":24})
ax3.set_title("Blue", fontdict={"fontsize":24})
ax4.set_title("Brightness", fontdict={"fontsize":24})
ax5.set_title("Grace Hopper", fontdict={"fontsize":34})
# create xticks for every histograms
ax1.set_xticks((0, 0.5, 1))
ax2.set_xticks((0, 0.5, 1))
ax3.set_xticks((0, 0.5, 1))
ax4.set_xticks((0, 0.5, 1))
# create ylim and yticks for every histograms
ax1.set_ylim((0,8000))
ax2.set_ylim((0,8000))
ax3.set_ylim((0,8000))
ax4.set_ylim((0,24000))

yti = [0, 4000, 8000]
ax1.set_yticks(yti)
ax1.set_yticklabels(yti)
ax2.set_yticks(yti)
ax2.set_yticklabels(yti)
ax3.set_yticks(yti)
ax3.set_yticklabels(yti)
# Bigger one for the brightness
ax4.set_yticks([0, 12000, 24000])
ax4.set_yticklabels([0, 12000, 24000])
# put the nickname of grace hooper below her picture
ax5.text(260, 560, "Amazing Grace", horizontalalignment='center', fontsize=20)

# separate the colors
red = imgArr[:,:,0]
green = imgArr[:,:,1]
blue = imgArr[:,:,2]
brightness = imgArr[:,:,0:3] 

# Give a title to those graphs
fig.suptitle('RGB SEPARATION', fontsize=38)

# Create an hist for each colors
ax1.hist(red.ravel(), histtype="stepfilled", bins=256, range=(0.0, 1.0), fc='r', ec='#d40000')
ax2.hist(green.ravel(), histtype="stepfilled", bins=256, range=(0.0, 1.0), fc='#00ba16', ec='g')
ax3.hist(blue.ravel(), histtype="stepfilled", bins=256, range=(0.0, 1.0), fc='#2067ff', ec='b')
ax4.hist(brightness.ravel(), histtype="stepfilled", bins=256, range=(0.0, 1.0), fc='#bdbdbd', ec='#545454')

plt.show()
```
