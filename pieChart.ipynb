```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 7))

options = [25, 15, 60, 40, 14]
optionsLabel = ["option1", "option2", "option3", "option4", "option5"]
colors = ['C3', 'C4', 'C0', 'C9', 'C6']
Xplode = [0, 0.1, 0, 0.1, 0]


ax.pie(options, colors=colors, labels=optionsLabel, shadow=True, autopct='%.1f%%', explode=Xplode,\
        startangle=90, counterclock=False, textprops={'fontsize': 18, 'weight': 'semibold'}, \
      wedgeprops = { 'linewidth' : 1.3 , 'edgecolor' : 'black'})

ax.axis('equal')

fig.suptitle('Pie Chart', fontsize=24, fontweight='bold')

plt.savefig('camembert.png', format='png')

plt.show()
```
